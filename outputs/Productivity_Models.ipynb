```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Model Comparison Table
results = pd.DataFrame({
    'Model': [
        'Linear Regression',
        'Ridge Regression',
        'Random Forest Regressor',
        'Gradient Boosting Regressor',
        'Support Vector Regressor'
    ],
    'R¬≤ Score': [0.91, 0.91, 0.93, 0.93, 0.90],
    'MAE': [0.43, 0.43, 0.40, 0.39, 0.43],
    'RMSE': [0.58, 0.58, 0.58, 0.58, 0.58]
})

# Display the table
print("üîç Model Performance Comparison:\n")
display(results.sort_values(by="R¬≤ Score", ascending=False))

# Step 2: Bar Plot of R¬≤ Score
plt.figure(figsize=(10, 6))
plt.bar(results['Model'], results['R¬≤ Score'], color='mediumseagreen')
plt.title(' R¬≤ Score Comparison Across Models', fontsize=14)
plt.ylabel('R¬≤ Score')
plt.xticks(rotation=30, ha='right')
plt.ylim(0.88, 0.95)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 3: Print Final Conclusion
best_model = results.sort_values(by="R¬≤ Score", ascending=False).iloc[0]['Model']

print("\n Conclusion:")
print(f"The model with the highest R¬≤ Score is **{best_model}**, achieving an R¬≤ of {results['R¬≤ Score'].max():.2f}.")
print("Both Gradient Boosting and Random Forest models performed best overall, likely due to their ability to capture non-linear patterns.")
print("Therefore, ensemble methods are recommended for predicting productivity scores based on behavioral and digital exposure features.")
```
